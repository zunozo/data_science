## 데이터프레임

- 데이터프레임에는 숫자, 문자열, 불리언 등 임의의 타입의 데이터를 담을 수 있다.

```python
import pandas as pd
import numpy as np
dic = {'city': ['서울', ' 부산', '대전' ,'대구' , '광주'],
 'year': [2017, 2017, 2018, 2018, 2018],
 'temp': [18, 20, 19, 21, 20]}
data = pd.DataFrame(dic)
print(data)
print(data[['year', 'city', 'temp']])
data.index = ['a','b','c','d','e']
data['year']
```

```python
cars = [50,40,20,30,10]
data['car'] = cars
data['high'] = data.temp >= 20 ; data
```

```python
data. drop(['car','high'],1)
```

### apply
- 함수를 편리하게 데이터에 일괄 적용하는 방법으로 apply가 있다.

```python
f = lambda x: x.max() - x.min()
df = pd.DataFrame(np.arange(12).reshape(4,3),columns=['A','B','C'],index=['a',
'b', 'c', 'd'])
print(df)
df.apply(f, axis=1)
```

### Series
- 판다스는 시리즈 객체를 제공하는데 이를 컬럼이 하나뿐인 데이터 프레임이다. 즉, 특수한 구조의 데이터프레임이라고 할 수 있다.

```python
region = pd.Series(['서울' , '부산', '대전' ,'대구' , '광주'],
index=['1','2','3','4','5'])
print(region)
```
